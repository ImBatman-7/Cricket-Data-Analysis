# process match summary

```python
import pandas as pd
import json
with open('t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)

data
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard' : 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
dict_match_id = {}
for row in df_match.itertuples():
    key1 , key2 = row.team1 + " Vs " + row.team2, row.team2 + " Vs " + row.team1
    dict_match_id[key1], dict_match_id[key2] = row.match_id, row.match_id

dict_match_id
```

# process batting summary

```python
with open('t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_summary = []
    for i in data:
        all_summary.extend(i['battingSummary'])
len(all_summary)
df_summary = pd.DataFrame(all_summary)
df_summary.tail(7)
```

```python

df_summary['dismissal'] = df_summary['dismissal'].apply(lambda x : "out" if len(x) > 0 else "not_out")
df_summary['batsmanName'] = df_summary['batsmanName'].apply(lambda x : x.replace('â€' , ''))
df_summary['batsmanName'] = df_summary['batsmanName'].apply(lambda x : x.replace('(c)' , ''))
df_summary['batsmanName'] = df_summary['batsmanName'].apply(lambda x : x.replace('\xa0' , ''))
df_summary.tail(10)
```

```python
df_summary["match_id"] = df_summary["match"].map(dict_match_id)
```

```python
df_summary.tail()
```

```python
df_temp = df_summary.copy()
df_temp.head()
```

```python
df_summary.head()
```

```python
df_summary.to_csv('temp.csv', index = False)
```
